```python
#Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading the .csv file and correction of typo in Salary column name
data = pd.read_csv('project3.csv')
data.rename(columns={'Brutto Salary montly': 'Monthly Brutto Salary'}, inplace=True)
data.head()
```

```python
#Checking of data types and missing values
data.info()
```

```python
#Correction of wrong display of salaries. Many job postings list an annual salary
#Thus i divided all salaries, which have value greater than 20k to the 14 

data['Monthly Brutto Salary'] = pd.to_numeric(data['Monthly Brutto Salary'], errors='coerce')

# Define a function to adjust salaries, considering NaN values
def adjust_salary(salary):
    if pd.isna(salary):
        return salary  
    elif salary > 20000:
        return round(salary / 14, 2)
    else:
        return salary

data['Monthly Brutto Salary'] = data['Monthly Brutto Salary'].apply(adjust_salary)

#Check the results
data.head() 
```

```python
#replace NaN values with "Not specified". Missing values in column Salary, could be handled better by using statistical methods.
#Despite this, I decided to leave them as they are described in job postings. 
data['Monthly Brutto Salary'] = data['Monthly Brutto Salary'].fillna('Not specified')
data['Technical Skills'] = data['Technical Skills'].fillna('Not specified')
data['Soft Skills'] = data['Soft Skills'].fillna('Not specified')
data['Branch'] = data['Branch'].fillna('Not specified')
data.head()
```

```python
#All NaN values have been handled 
data.info()
```

```python
#Handling some problems with Tech Skills column. Because the parser didn't work perfectly, there are some mistakes in the cells like: 
#typos, words with commas without space. Also the main problem is, that there exists many synonymous expressions like "analytical thniking" and "logical thinking"
#So i decided to replace them with identical ones

data['Tech Skills List'] = data['Technical Skills'].str.split(', ')

all_tskills = data['Tech Skills List'].explode()

all_tskills_replaced = all_tskills.replace(['PowerBI', 'MS Power BI'], 'Power BI')
all_tskills_replaced = all_tskills_replaced.replace('RStudio', 'R')
all_tskills_replaced = all_tskills_replaced.replace([' MySQL', ], 'SQL')
all_tskills_replaced = all_tskills_replaced.replace('SAP Business Objects', 'SAP')
all_tskills_replaced = all_tskills_replaced.replace(['Cloud', 'Cloud-Technologien'], 'Cloud technologies')
all_tskills_replaced = all_tskills_replaced.replace(['Azure ML'], 'Azure')

tskill_counts = all_tskills_replaced.value_counts()

tskill_counts_df = tskill_counts.reset_index()
tskill_counts_df.columns = ['Technical Skill', 'Count']

# Export to Excel
tskill_counts_df.to_excel('technical_skills_counts.xlsx', index=False)

tskill_counts.head(8)


```

```python
data['Soft Skills List'] = data['Soft Skills'].str.split(', ')

all_skills = data['Soft Skills List'].explode()

all_skills_replaced = all_skills.replace(['communication', 'communiaction', 'good communication', 'Excellent communication'], 'Communication Skills')
all_skills_replaced = all_skills_replaced.replace(['Analytical', 'analytical', 'analytical thinking', 'analytica thinking',
                                                   'analytical skills', 'logical thinking', 'critical thinking', 'networked thinking', ], 'Analytical thinking')
all_skills_replaced = all_skills_replaced.replace(['team Player', 'teamwork', 'team player'], 'Team Player')
all_skills_replaced = all_skills_replaced.replace(['work independently', 'Independent work', 'independent working style', 'independent', 'Independent work',
                                                   'independent working style'], 'Independent working style')
all_skills_replaced = all_skills_replaced.replace(['creativity', 'creative'], 'Creativity')
all_skills_replaced = all_skills_replaced.replace(['structured', 'and structured way of working'], 'Structured way of working')
all_skills_replaced = all_skills_replaced.replace(['organizational skills', 'organizational talent'], 'Organizational Skills')
all_skills_replaced = all_skills_replaced.replace(['interpersonal skills', 'interpersonal'], 'Interpersonal Skills')
all_skills_replaced = all_skills_replaced.replace(['presentation', 'public presentations'], 'Interpersonal Skills')
all_skills_replaced = all_skills_replaced.replace(['initiative', 'Personal initiative'], 'Initiative')



skill_counts = all_skills_replaced.value_counts()

skill_counts_df = skill_counts.reset_index()
skill_counts_df.columns = ['Soft Skill', 'Count']

# Export to Excel
skill_counts_df.to_excel('soft_skills_counts.xlsx', index=False)

skill_counts.head(10)

```
